### 1. 定义队列类对象

```python
class SequenceDoubleQueue():
    # 初始化
    def __init__(self):
        self.array = []
    
    # 判断队列是否为空
    def QueueEmpty(self):
        return self.array == []
    
    # 返回队头元素
    def GetHead(self):
        if self.QueueEmpty():
            print("队列为空！")
            return
        return self.array[0]
    
    # 返回队列的长度
    def QueueLength(self):
        return len(self.array)
    
    # 队头入队：添加元素到队头
    def AddHead(self, item):
        self.array.insert(0, item)
    
    # 队尾入队：添加元素到队尾
    def AddTail(self, item):
        self.array.append(item)
    
    # 队头出队：删除队头元素
    def DropHead(self):
        self.array.pop(0) # pop(0):删除第一个元素
    
    # 队尾出队：删除队尾元素
    def DropTail(self):
        self.array.pop() # 删除最后一个元素
    
    # 遍历队列
    def Traverse(self):
        for item in self.array:
            print(item, end=' | ')
    
    # 清空队列
    def clear(self):
        self.array = []
        print("清空队列完成！")
```

### 2. 创建一个队列对象

```python
import random
random.seed(10)
items = [random.randint(1, 100) for i in range(5)]
queue = SequenceDoubleQueue()
print(queue)
```

### 3. 判断队列是否为空

```python
if queue.QueueEmpty():
    print("队列为空！")
else:
    print("队列不为空！")
```

### 4. 队头入队：添加元素到队头

```python
for item in items:
    queue.AddHead(item)
queue.Traverse() # 遍历队列
```

### 5. 队尾入队：添加元素到队尾

```python
for item in items:
    queue.AddTail(item)
queue.Traverse() # 遍历队列
```

### 6. 队头出队：删除队头元素

```python
print("删除前的队列:", end=" ")
queue.Traverse()
queue.DropHead() # 出队
print()
print("删除后的队列:", end=" ")
queue.Traverse() # 遍历队列
```

### 7. 队尾出队：删除队尾元素

```python
print("删除前的队列:", end=" ")
queue.Traverse()
queue.DropTail() # 出队
print()
print("删除后的队列:", end=" ")
queue.Traverse() # 遍历队列
```

### 8. 输出队列长度

```python
print("队列长度为：", queue.QueueLength())
```

### 9. 清空队列

```python
queue.clear()
```
