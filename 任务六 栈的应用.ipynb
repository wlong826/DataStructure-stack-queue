### 1. 定义类对象栈

```python
class Stack():
    # 初始化栈
    def __init__(self):
        # 栈的容器采用顺序表的形式实现，顺序表的尾部操作复杂度为O(1)，所以添加元素和去除元素在尾部操作耗时少，单链表反之
        self.items = [] # 创建空栈，这里用数组来表示
        
    # 判断是否为空
    def StackEmpty(self):
        return self.items == []
    
    # 返回栈顶元素
    def GetTop(self):
        return self.items[-1]
    
    # 返回栈的长度
    def StackLength(self):
        return len(self.items)
    
    # 入栈
    def Push(self, item):
        self.items.append(item)
    
    # 出栈
    def Pop(self):
        return self.items.pop()
    
    # 遍历
    def StackTraverse(self):
        for item in items:
            print(item, end = '|')
    
    # 清空栈
    def Clear(self):
        self.items = []
```

### 2. 应用1：括号匹配问题

　　检验算法借助一个栈，每当读入一个左括号，则直接入栈，等待相匹配的同类右括号；每当读入一个右括号，若与当前栈顶的左括号类型相同，则二者匹配，将栈顶的左括号出栈，直到表达式扫描完毕。

　　在处理过程中，还要考虑括号不匹配出错的情况。例如，当出现 (()[]) 这种情况时，由于前面入栈的左括号均已知和后面出现的右括号相匹配，栈已空，因此最后扫描的右括号不能得到匹配；出现 [([]) 这种错误，当表达式扫描结束时，栈中还有一个左括号没有匹配；出现 (()] 这种错误显然是栈顶的左括号和最后的右括号不匹配。

**测试案例**：（小伙伴们需要拿以下三个案例测试自己写的代码哈！）
　　案例1：{{([][])}()}
　　案例2：[[{{(())}}]]
　　案例3：[[(()(()))])]{}

```python
# 匹配函数，最终返回True和False
def matching(strings):
    stack = Stack() # 创建一个空栈，存储左括号
    flag = 1 # 标记匹配结果以控制循环及返回结果
    opens = '{[(' 
    closes = '}])' # 必须和opens里的括号相匹配
    for string in strings:
        if string in opens:
            stack.Push(string) # 当为左括号时，入栈
        elif string in closes:
            ele = stack.Pop() # 当为右括号时，将最后一个元素出栈
            if opens.index(ele) != closes.index(string): # 如果出栈的元素和右括号位置不匹配，则匹配错误！
                return False
        else:
            return False
    return stack.StackEmpty()  # 根据最后的元素是否为空，判断是否匹配正确

# 判断返回的是True还是False
def check(strings):
    if matching(strings):
        print("%s 匹配正确!" % strings)
    else:
        print("%s 匹配错误！" % strings)
```

### 3. 应用2：十进制转二进制

　　当将一个十进制整数M转换为二进制数时，在计算过程中，把M与2求余得到的二进制数的各位依次进栈，计算完毕后将栈中的二进制数依次出栈输出，输出结果就是待求得的二进制数。

**测试案例**：
　　[200, 254, 153, 29, 108, 631, 892]
	
**运行结果**：

![Image Name](https://cdn.kesci.com/upload/image/qw9o9hwjth.png?imageView2/0/w/320/h/320)

　【注】：小伙伴们需要完成以上测试哈，而且必须使用哦，主要为了方便作业统一！

```python
# 利用栈实现十进制转二进制
def divideBy2(number):
    stack = Stack() #创建一个空栈，存储二进制01数
    
    while number > 0:
        mod = number % 2 # number对2求模，得到0或者1
        stack.Push(mod)
        number = number // 2 # number整除2 求商
    
    # 定义一个存储二进制数的空字符串
    binString = ''
    while not stack.StackEmpty():
        binString = binString + str(stack.Pop()) # 调用Pop删除栈顶元素并返回值
    
    return binString
```

```python
# 测试应用2
numbers = [200, 254, 153, 29, 108, 631, 892]
for number in numbers:
    print("%d 的二进制为：" % number, divideBy2(number))
```

### 4. 应用3：十进制转N进制（应用2的拓展）

　　当将一个十进制整数M转换为N进制数时，在计算过程中，把M与N求余得到的N进制数的各位依次进栈，计算完毕后将栈中的N进制数依次出栈输出，输出结果就是待求得的N进制数。

**测试案例**：
　　测试数：[200, 254, 153, 29, 108, 631, 892]
　　进制：[4, 8, 16]
	
　　自由搭配即可，这里我主要使用了**random.choice()**来对测试数要转换的进制随机取数，测试代码如下所示，仅供参考；

```python
# 十进制转四进制、八进制、十六进制
def divideByN(number, base): # base代表几进制
    digits = '0123456789ABCDEF' # 由于十六进制后面是用字母代替，所以这里digits主要为了转换
    stack = Stack() # 创建一个空栈，存储N进制数
    
    while number > 0:
        mod = number % base # number对base求模
        stack.Push(mod) # 将余数（0到base之间的数）进栈
        number = number // base
    
    baseString = '' # 创建一个空字符串，存放N进制数
    while not stack.StackEmpty():
        baseString = baseString + digits[stack.Pop()] # 调用Pop()函数弹出栈顶元素
        
    return baseString # 返回值
```

```python
import random
numbers = [200, 254, 153, 29, 108, 631, 892]
bases = [4, 8, 16]
for number in numbers:
    base = random.choice(bases)
    print("%d 的 %d 进制数为：%s" % (number, base, divideByN(number, base)))
```
