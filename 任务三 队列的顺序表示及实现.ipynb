### 1. 定义队列类

```python
class Queue():
    # 初始化
    def __init__(self):
        self.array = []
    
    # 判断队列是否为空
    def QueueEmpty(self):
        return self.array == []
    
    # 返回队头元素
    def GetHead(self):
        if self.QueueEmpty():
            print("队列为空！")
            return
        return self.array[0]
    
    # 返回队列的长度
    def QueueLength(self):
        return len(self.array)
    
    # 入队，添加元素到队尾
    def EnQueue(self, item):
        self.array.append(item)
    
    # 出队，删除队头元素
    def DeQueue(self):
        self.array.pop(0) # pop(0):删除第一个元素
        
    # 遍历队列
    def Traverse(self):
        for item in self.array:
            print(item, end=' | ')
    
    # 清空队列
    def clear(self):
        self.array = []
        print("清空队列完成！")
```

### 2. 创建一个队列对象

```python
import random
random.seed(10)
items = [random.randint(1, 100) for i in range(10)]
queue = Queue()
print(queue)
```

### 3. 判断队列是否为空

```python
if queue.QueueEmpty():
    print("队列为空！")
else:
    print("队列不为空！")
```

### 4. 入队：向队尾添加元素

```python
for item in items:
    queue.EnQueue(item)
queue.Traverse() # 遍历队列
```

### 5. 出队：删除队首元素

```python
print("删除前的队列:", end=" ")
queue.Traverse()
queue.DeQueue() # 出队
print()
print("删除后的队列:", end=" ")
queue.Traverse() # 遍历队列
```

### 6. 输出队列长度

```python
print("队列长度为：", queue.QueueLength())
```

### 7. 清空队列

```python
queue.clear()
```
