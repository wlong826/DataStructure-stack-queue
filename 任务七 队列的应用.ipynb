### 1. 定义队列类对象

```python
class SequenceDoubleQueue():
    # 初始化
    def __init__(self):
        self.array = []
    
    # 判断队列是否为空
    def QueueEmpty(self):
        return self.array == []
    
    # 返回队头元素
    def GetHead(self):
        if self.QueueEmpty():
            print("队列为空！")
            return
        return self.array[0]
    
    # 返回队列的长度
    def QueueLength(self):
        return len(self.array)
    
    # 队头入队：添加元素到队头
    def AddHead(self, item):
        self.array.insert(0, item)
    
    # 队尾入队：添加元素到队尾
    def AddTail(self, item):
        self.array.append(item)
    
    # 队头出队：删除队头元素
    def DropHead(self):
        return self.array.pop(0) # pop(0):删除第一个元素
    
    # 队尾出队：删除队尾元素
    def DropTail(self):
        return self.array.pop() # 删除最后一个元素
    
    # 遍历队列
    def Traverse(self):
        for item in self.array:
            print(item, end=' | ')
    
    # 清空队列
    def clear(self):
        self.array = []
        print("清空队列完成！")
```

### 2. 应用1：回文词检验

　　回文词检查是数据结构中的常见任务，这一任务可以应用双端队列直观有效地完成。将待检查的字符串按顺序载入队列，首尾两侧弹出并比较，出现不一致时则判断为不是回文数。当队列为空或只剩一个元素是可以判断为回文数。

**测试案例**：
　　案例1：abcgcba
　　案例2：refer
　　案例3：reference
	
**运行结果**：

```python
# 回文词判定
def palchecker(aString): # 定义回文词判断函数，需要一个参数字符串 aString
    queue = SequenceDoubleQueue()  # 初始化一个双端队列 queue
    
    for ch in aString:
        queue.AddTail(ch) # 依次双端队列尾端加入所有aString的元素
        
    label = True # 
    
    while queue.QueueLength() > 1 and label:  # 开始循环的查找 当双端队列的长度大于1并且仍然相等
        first = queue.DropHead()  # first取出双端队列中的首端数据
        last = queue.DropTail()    # last取出双端队列中的尾端数据
        if first != last:    # 进行比较 如果不相等
            label = False   # label 取False 说明不相等，不平衡，不是回文数
            
    return label  # 返回标志符label label为True说明是回文数
```

```python
## 测试类代码
aStrings = ["abcgcba", "refer", "reference"]
for aString in aStrings:
    if palchecker(aString):   # 与奇数偶数无关
        print("%s 是回文词！" % aString)
    else:
        print("%s 不是回文词！" % aString)
```
