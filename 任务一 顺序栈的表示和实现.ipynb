### 1. 定义栈类对象

```python
class Stack():
    # 初始化栈
    def __init__(self):
        # 栈的容器采用顺序表的形式实现，顺序表的尾部操作复杂度为O(1)，所以添加元素和去除元素在尾部操作耗时少，单链表反之
        self.items = [] # 创建空栈，这里用数组来表示
        
    # 判断是否为空
    def is_empty(self):
        return self.items == []
    
    # 返回栈顶元素
    def top(self):
        return self.items[-1]
    
    # 返回栈的长度
    def length(self):
        return len(self.items)
    
    # 压栈、入栈、进栈
    def push(self, item):
        self.items.append(item)
    
    # 出栈
    def pop(self):
        return self.items.pop()
    
    # 清空栈
    def clear(self):
        self.items = []
        print("当前栈已清空！")
```

### 2. 创建顺序栈对象

```python
import random
random.seed(10)
items = [random.randint(1,100) for i in range(10)]
# 创建一个栈对象
stack = Stack()
print(stack)
```

### 3. 判断栈是否为空

```python
if stack.is_empty():
    print("栈为空！")
else:
    print("栈不为空！")
```

### 4. 将元素入栈

```python
for item in items:
    stack.push(item) # 进栈
```

### 5. 将最后一个元素出栈

```python
print("出栈的元素为:", stack.pop())
```

### 6. 返回栈顶元素

```python
print("栈顶元素为：", stack.top())
```

### 7. 返回栈的长度

```python
print("栈的长度为:", stack.length())
```

### 8. 清空栈

```python
stack.clear()
```
