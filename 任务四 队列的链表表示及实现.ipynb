### 1. 定义队列类对象

```python
# 队列结点类
class QueueNode():
    # 结点类的初始化
    def __init__(self, item):
        self.item = item
        self.next = None
# 队列类
class Queue():
    # 初始化
    def __init__(self):
        self.head = None
    
    # 判断队列是否为空
    def QueueEmpty(self):
        return self.head == None
    
    # 返回队头元素
    def GetHead(self):
        if self.QueueEmpty():
            print("队列为空！")
            return
        return self.head
    
    # 返回队列的长度
    def QueueLength(self):
        return len(self.Traverse())
    
    # 入队，添加元素到队尾
    def EnQueue(self, item):
        node = QueueNode(item)
        if self.QueueEmpty():
            self.head = node
            return
        nod = self.head
        while nod.next != None:
            nod = nod.next
        nod.next = node
    
    # 出队，删除队头元素
    def DeQueue(self):
        nod = self.head
        self.head = nod.next
    
    # 遍历队列
    def Traverse(self):
        if self.QueueEmpty():
            print("队列为空！")
            return
        show_list = []
        nod = self.head
        while nod != None:
            show_list.append(nod.item)
            nod = nod.next
        return show_list
    
    # 清空队列
    def clear(self):
        self.head = None
        print("清空队列完成！")
```

### 2. 创建一个队列对象

```python
import random
random.seed(10)
items = [random.randint(1, 100) for i in range(10)]
queue = Queue()
print(queue)
```

### 3. 判断队列是否为空

```python
if queue.QueueEmpty():
    print("队列为空！")
else:
    print("队列不为空！")
```

### 4. 入队：向队尾添加元素

```python
for item in items:
    queue.EnQueue(item)
queue.Traverse() # 遍历队列
```

### 5. 出队：删除最后一个元素

```python
print("删除前的队列:", queue.Traverse())
queue.DeQueue() # 出队
print("删除后的队列:", queue.Traverse()) # 遍历队列
```

### 6. 输出队列长度

```python
print("队列长度为：", queue.QueueLength())
```

### 7. 清空队列

```python
queue.clear()
```
