### 1. 定义栈类对象

```python
# 结点类，包含两个元素，一个是数据域、一个是指向下一节点的next指针;
class StackNode():
    def __init__(self, item):
        self.item = item
        self.next = None 
# 栈类
class Stack():
    # 初始化栈
    def __init__(self):
        self.top = None # 栈顶
        self.head = None # 栈底
        self.size = 0 # 表示链的长度
        
    # 判断是否为空
    def is_empty(self):
        return self.size == 0
    
    # 返回栈顶元素
    def getTop(self):
        if self.is_empty():
            print("空栈！")
            return
        return self.top.item
    
    # 返回栈的长度
    def get_length(self):
        return self.size
    
    # 遍历栈
    def travel(self):
        if self.is_empty():
            print("空栈！")
            return
        show_list = []
        nod = self.head
        while nod != None:
            show_list.append(nod.item)
            nod = nod.next
        return show_list
    
    # 压栈、入栈、进栈
    def push(self, item):
        node = StackNode(item)
        if self.is_empty():
            self.top = node
            self.head = node
            self.size += 1
            return
        self.top.next = node
        self.top = node
        self.size += 1
    
    # 出栈
    def pop(self):
        if self.is_empty():
            print("栈为空栈！")
            return
        self.size -= 1
        item = self.top.item
        nod = self.head
        while nod.next != self.top:
            nod = nod.next
        nod.next = None
        self.top = nod
        return item
    
    # 清空栈
    def clear(self):
        self.head = None # 令栈底为空
        self.top = None # 令栈顶为空
        self.size = 0 # 令栈大小为0
        print("当前栈已清空！")
```

### 2. 创建顺序栈对象

```python
import random
random.seed(10)
items = [random.randint(1,100) for i in range(10)] # 定义一个长度为10的列表，为了方便进栈
# 创建一个栈对象
stack = Stack()
print(stack)
```

### 3. 判断栈是否为空

```python
if stack.is_empty():
    print("栈为空！")
else:
    print("栈不为空！")
```

### 4. 元素入栈

```python
for item in items:
    stack.push(item) # 进栈
print("当前栈为：", stack.travel())
```

### 5. 元素出栈

```python
print("出栈的元素为:", stack.pop())
```

### 6. 返回栈顶元素

```python
print("当前栈为：", stack.travel())
print("栈顶元素为：", stack.getTop())
```

### 7. 返回栈的长度

```python
print("栈的长度为:", stack.get_length())
```

### 8. 清空栈

```python
stack.clear()
```
